```python
import matplotlib.pyplot as plt

# Specify the file path
file_path = "data/strong_scaling_atomic_result3.txt"

# Read data from the file
with open(file_path, 'r') as file:
    data = file.read()

# Split the data into lines and then split each line into thread and time values
lines = data.split('\n')

# Initialize empty lists for threads and times
threads = []
times = []

# Process each line
for line in lines:
    # Skip empty lines
    if not line:
        continue

    # Split the line into thread and time values
    values = line.split()

    # Check if the line contains at least two values
    if len(values) >= 2:
        thread, time = values[:2]
        threads.append(int(thread))
        times.append(float(time))

# Calculate speedup
speedup = [times[0] / time for time in times]

# Plot the data
plt.plot(threads, speedup, marker='o')
plt.title('Speedup vs Threads for strong scaling with sum as atomic')
plt.xlabel('Threads')
plt.ylabel('Speedup')
plt.grid(True)
plt.show()
```

```python
import os
import matplotlib.pyplot as plt

# Specify the file path
file_path = os.path.join("data", "bonus_b_acutualerror_esterror.dat")

# Read data from the file
with open(file_path, 'r') as file:
    data = file.read()

# Parse the data
lines = data.split('\n')
N_values = []
actual_estimate_values = []
estimated_error_values = []

for line in lines:
    parts = line.split()
    if len(parts) == 4:  # Ensure that the line has enough parts to extract the required values
        N_values.append(int(parts[0].split('=')[1]))
        actual_estimate_values.append(float(parts[2].split('=')[1]))
        estimated_error_values.append(float(parts[3].split('=')[1]))
# Plot actual_estimate vs N
plt.figure(figsize=(10, 5))
plt.loglog(N_values, actual_estimate_values, marker='o', label='Actual Error(2^10-Integral)')
plt.title('Actual Error vs N')
plt.xlabel('N')
plt.ylabel('Actual Error')
plt.grid(True)
plt.legend()
plt.show()

# Plot estimated_error vs N
plt.figure(figsize=(10, 5))
plt.loglog(N_values, estimated_error_values, marker='o', color='orange', label='Estimated Error(previousIntegral-currentIntegral)')
plt.title('Estimated Error vs N')
plt.xlabel('N')
plt.ylabel('Estimated Error')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import os
import matplotlib.pyplot as plt

# Specify the file path
file_path = os.path.join("data", "3c_anvil_data.dat")

# Read data from the file
with open(file_path, 'r') as file:
    lines = file.readlines()

# Parse the data
N_values, integral_values = zip(*(map(float, line.split()) for line in lines))

# Plot log-log plot
plt.figure(figsize=(10, 5))
plt.loglog(N_values, integral_values, marker='o')
plt.title('Log-Log Plot of N vs Integral')
plt.xlabel('N')
plt.ylabel('Integral')
plt.grid(True)
plt.show()
```
